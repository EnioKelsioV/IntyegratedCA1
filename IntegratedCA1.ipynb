# Data Preparation

## Dataset Overview

### The dataset contains 19,075 rows and 8 columns, detailing food imports:

	1.	Commodity – Type of imported food (e.g., “Total live animals”)
    2.	Country – The country of import origin
	3.	Unit of Measure – The measurement unit (e.g., “Million $”, “1,000”).
	4.	Category – General category of the product
	5.	Subcategory – A more specific classification within the category (e.g., “Foods”).
	6.	Row Number – A sequential identifier.
	7.	Year Number – The year of the import transaction.
	8.	Value – The numerical value of the import (in specified units).

# Data Visualisation

##  Import Libraries and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "FoodImports.csv" 
df = pd.read_csv(file_path, encoding="ISO-8859-1")
```

## Plot the Top 10 Imported Commodities

```python
plt.figure(figsize=(12, 6))
top_commodities = df.groupby("Commodity")["Value"].sum().nlargest(10)
top_commodities.plot(kind="bar", color="teal")

plt.title("Top 10 Imported Commodities by Value", fontsize=14)
plt.xlabel("Commodity", fontsize=12)
plt.ylabel("Total Import Value", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

## Visualize Import Trends Over the Years

```python
plt.figure(figsize=(12, 6))
df_grouped = df.groupby("Year Number")["Value"].sum()
df_grouped.plot(marker="o", linestyle="-", color="royalblue")

plt.title("Total Food Imports Over the Years", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Import Value", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
